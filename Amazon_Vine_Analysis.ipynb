```python
# Start Spark session
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("VineReviewAnalysis").getOrCreate()
```

```python
# There is a DataFrame or table for the vine_table data using one of three methods above
from pyspark import SparkFiles
url = "https://s3.amazonaws.com/amazon-reviews-pds/tsv/amazon_reviews_us_Watches_v1_00.tsv.gz"
spark.sparkContext.addFile(url)
watches_df = spark.read.option("encoding", "UTF-8").csv(SparkFiles.get("amazon_reviews_us_Watches_v1_00.tsv.gz"), sep="\t", header=True, inferSchema=True)

# Row Count
watches_df.count()
```

```python
from pyspark.sql.functions import col, avg

# Filter by voters.
filtered_watches_df = watches_df.select(["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"])
filtered_watches_df.show(10)
```

```python
# The data is filtered to create a DataFrame or table where there are 20 or more total votes (5 pt)
twenty_or_more_df = filtered_watches_df.filter(filtered_watches_df['total_votes'] >= 20)
twenty_or_more_df.show(10)
```

```python
# The data is filtered to create a DataFrame or table where the percentage of helpful_votes is equal to or greater than 50% (5 pt)
fifty_percent_df = twenty_or_more_df.filter(twenty_or_more_df["helpful_votes"]/twenty_or_more_df["total_votes"] >= 0.5)
fifty_percent_df.show(10)
```

```python
# The data is filtered to create a DataFrame or table where there is a Vine review (5 pt)
vine_review_df = fifty_percent_df.filter(fifty_percent_df['vine']== 'Y')
vine_review_df.show(10)
```

```python
# The data is filtered to create a DataFrame or table where there isnâ€™t a Vine review (5 pt)
no_vine_review_df = fifty_percent_df.filter(fifty_percent_df['vine']== 'N')
no_vine_review_df.show(10)
```

```python
# The total number of reviews, the number of 5-star reviews, and the percentage 5-star reviews are calculated for all Vine and non-Vine reviews (15 pt)

# Determine the percentage of five-star reviews among Vine reviews
vine_number = vine_review_df.count()
vine_five_star = vine_review_df.filter(vine_review_df['star_rating']== 5).count()
percentage_vine_five_star = float(vine_five_star) / float(vine_number)
print(vine_number)
print(vine_five_star)
print(percentage_vine_five_star)

# Determine the percentage of five-star reviews among non-Vine reviews.
no_vine_number = no_vine_review_df.count()
no_vine_five_star = no_vine_review_df.filter(no_vine_review_df['star_rating']== 5).count()
percentage_no_vine_five_star = float(no_vine_five_star) / float(no_vine_number)
print(no_vine_number)
print(no_vine_five_star)
print(percentage_no_vine_five_star)

# Total amount of reviews
print(vine_number + no_vine_number)
```
